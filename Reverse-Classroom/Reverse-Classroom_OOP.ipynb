## Cosa é l'OOP?
OOP sta per Object Oriented Programming, ovvero programmazione orientata agli oggetti. **Esso è un paradigma della programmazione sviluppato e pensato per organizzare concetti (codice), un pò come il pensiero umano organizza le informazioni**

L'OOP non è un paradigma che si applica soltanto a Python. La maggior parte dei linguaggi di programmazione utilizzati tutt'oggi sfruttano questo concetto.

## Cenni storici
Il primo linguaggio di programmazione orientato agli oggetti fu il Simula (1967). Negli anni ottanta sono state create estensioni orientate ad oggetti del linguaggio C (C++, Objective-C, e altri). Negli anni novanta è diventato il paradigma dominante della programmazione, per cui gran parte dei linguaggi erano o nativamente orientati agli oggetti o avevano una estensione in tal senso. **Linguaggi che supportano solo il paradigma OOP sono Smalltalk ed Eiffel**. Più spesso si incontra una realizzazione non esclusiva del paradigma di programmazione orientata agli oggetti, come in C++, Java, Delphi, Python, C#, Visual Basic .NET, Perl, PHP (a partire dalla versione 4).

Il concetto di **classe** può essere considerato l'erede del tipo di dato astratto.
I costrutti sintattici che permettono di definire una classe, nei linguaggi a oggetti, possono essere visti come un supporto strutturato per realizzare i dati astratti.

La programmazione ad oggetti prevede di raggruppare in una zona circoscritta del codice sorgente (chiamata classe), la dichiarazione delle strutture dati e delle procedure che operano su di esse. Le classi, quindi, costituiscono dei modelli astratti, che a tempo di esecuzione vengono invocate per istanziare o creare oggetti software relativi alla classe invocata. Questi ultimi sono dotati di attributi (dati) e metodi (procedure) secondo quanto definito/dichiarato dalle rispettive classi.

## La Classe
Quando andremo a creare questi oggetti, queste istanze contenenti la nostra porzione di codice, dobbiamo essenzialmente ***definire la forma di questo oggetto***. 
Immaginiamo di dover preparare dei biscotti, di diverse forme e dimensioni, ecco definire la “forma” di questo oggetto vuol dire costruire e progettare gli stampini di questi biscotti.
Tornando nel contesto tecnico Python, questi stampini vengono chiamati `Classi` e sono un modello, un template, un progetto di partenza che determinerà **caratteristiche (attributi)** e **azioni (metodi)** di ogni biscotto che andremo a creare.

Qui per esempio vediamo un esempio di come definire la nostra classe Dog(), il più semplicemente possibile.

Qui per esempio vediamo un esempio di come definire la nostra classe Dog(), il più semplicemente possibile.

```python
class Dog:
    def __init__(self):
        print("Bau!")
```

Ogni volta che andremo a creare una nuova istanza, in automatico Python andrà a cercare il metodo speciale `__init__` e lo andrà ad eseguire. Questo metodo ha lo scopo di inizializzare l’oggetto con quello che andremo a passare come “parametro”.

Attenzione, il suddetto metodo è opzionale non è obbligatorio. Tuttavia, se presente, Python lancerà questa funzione speciale ogni volta che l'oggetto Class verrà istanziato.

Il parametro **self** è un riferimento all'istanza che andremo a creare con la classe che abbiamo definito e viene generalmente utilizzato per accedere alle variabili o metodi che appartengono alla classe.

Non deve essere denominato necessariamente `"self"` , è possibile chiamarlo come si desidera, ma deve essere il primo parametro di qualsiasi funzione nella classe:

```python
bobby = Dog()
```

Ora vediamo più nello specifico come potremmo usare le classi. Supponiamo di dover rappresentare uno zoo, e di conseguenza dover definire una classe Pet che rappresenti tutti gli animali in esso contenuti.

```python
class Pet:
    def __init__(self, name, species):
        self.name = name
        self.species = species
        print(f"Hello! my name is: {self.name} and i am a: {self.species}")
```

Definita la nostra classe con i primi due attributi `name` e `species` andiamo a inizializzare le istanze.

```python
blue_gir = Pet("Blue", "Giraffe")
wyatt_ele = Pet("Wyatt", "Elephant")
joomla_tig = Pet("Joomla", "Tiger")
```
